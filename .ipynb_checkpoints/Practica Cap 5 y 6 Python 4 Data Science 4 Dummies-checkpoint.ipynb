## Muestreando datos de otras maneras
En este primer ejemplo se utiliza la función "enumerate" para crear precisamente una lista o enumeración de las lineas del archivo. Esto permite cargar en j el # de linea. 
En el ejemplo se utiliza el operador % (conocido como modulo o resudio) para solo hacer print de las líneas pares.

```python
n = 2
with open("Colors.txt", 'r') as open_file:
    for j, observation in enumerate(open_file):
        if j % n==0:
            print('Reading Line: ' + str(j) + ' Content: ' + observation)
```

Este caso hace una lectura random de lineas. Note que no es que por sí mismo hagan una lectura al azar de las lineas. Se cargan secuencialmente pero mediante condicional se determina cuáles pasan y cuáles no. Se puede ejecutar N cantidad de veces y en todos los casos se obtendrán resultados diferentes. 

En comentarios se deja la función seed la cual permitirá obtener siempre el mismo resultado. 

```python
from random import random
sample_size = 0.25
## from random import seed
## seed(123)
with open("Colors.txt", 'r') as open_file:
    for j, observation in enumerate(open_file):
        if random()<=sample_size:
            print('Reading Line: ' + str(j) + ' Content: ' + observation)
```

### Datos NO Estructurados
En este ejemplo, example_file es una variable que cntiene un string. imread es una función para cargar una imagen. imshow prepara la imagen para plotear. show plotea la imagen. Nótese que imread y imshow tienen parametrizado mostrar la imagen en escala de grises

```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm

example_file = ("http://upload.wikimedia.org/wikipedia/commons/7/7d/Dog_face.png")
image = imread(example_file, as_gray=True)
plt.imshow(image, cmap=cm.gray)
plt.show() 
```

```python
type(image)
```

```python
image.shape
```

```python
print("data type: %s, shape: %s" % (type(image), image.shape))
```

El siguiente ejemplo muestra como la imagen al final es un array de bytes por lo tanto se puede subdimensionar

```python
image2 = image[10:70,5:70]
plt.imshow(image2, cmap=cm.gray)
plt.show()
```

Una imagen se puede redimensionar con la función resize. 

```python
image3 = resize(image, (30, 30), mode='symmetric')
plt.imshow(image3, cmap=cm.gray)
print("data type: %s, shape: %s" % (type(image3), image3.shape))
```

Las imágenes anteriores son arreglos de 2 dimensiones (ver cambio en método .shape de las imágenes). Si se desea convertir un arreglo bidimensional en un arreglo simple de bytes se utiliza la función flatten.

```python
image_row = image3.flatten()
print("data type: %s, shape: %s" % (type(image_row), image_row.shape))
print(image_row)
```

### Acceder datos de la web
En este ejemplo se lee un archivo local XML

```python
from lxml import objectify
import pandas as pd

xml = objectify.parse(open('XMLData.xml'))
root = xml.getroot()

df = pd.DataFrame(columns=('Number', 'String', 'Boolean'))

for i in range(0,4):
    obj = root.getchildren()[i].getchildren()
    row = dict(zip(['Number', 'String', 'Boolean'],
                   [obj[0].text, obj[1].text,
                    obj[2].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = df.append(row_s)

print(df)
```
